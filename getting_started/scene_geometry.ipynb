```python
%load_ext eradiate
import eradiate
from eradiate import unit_registry as ureg
eradiate.set_mode("mono")
```

```python
exp = eradiate.experiments.OneDimExperiment(
    measures={
        "type": "perspective",
        "origin": [-1000, 0, 110] * ureg.km,
        "target": [0, 0, 110] * ureg.km,
        "up": [0, 0, 1],
        "film_resolution": (320, 240),
    },
    atmosphere={
        "type": "homogeneous",
        "top": 100 * ureg.km,
    },
)
```

```python
result = eradiate.run(exp)
result.radiance.squeeze().plot.imshow(yincrease=False)
```
